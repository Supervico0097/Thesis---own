```python
import numpy as np
import pandas as pd
from random import randint

d = pd.Series(np.random.choice([randint(15,39), randint(50,60),randint(60,85)], p=[0.60, 0.30, 0.10]) for _ in range(10000))

d.plot.hist(bins=50)
# d
```
